```python
import pandas as pd

# Read the data
tags = pd.read_csv('/Users/siyaksares/Developer/GitHub/ReadersRadarProject/tags.csv')
book_tags = pd.read_csv('/Users/siyaksares/Developer/GitHub/ReadersRadarProject/book_tags.csv')
books = pd.read_csv('/Users/siyaksares/Developer/GitHub/ReadersRadarProject/books.csv')

# Convert tag names to lowercase for case-insensitive comparison
tags['tag_name'] = tags['tag_name'].str.lower()

# Remove special characters and variations in tag names
tags['tag_name'] = tags['tag_name'].str.replace(r'[^a-zA-Z0-9\s]', '')

# Convert tag_id to numeric (if possible)
tags['tag_id'] = pd.to_numeric(tags['tag_id'], errors='coerce')
book_tags['tag_id'] = pd.to_numeric(book_tags['tag_id'], errors='coerce')

# Drop rows with NaN tag_id
tags.dropna(subset=['tag_id'], inplace=True)
book_tags.dropna(subset=['tag_id'], inplace=True)

# Get the top 10 most used tags
top_tags = book_tags.groupby('tag_id').count().sort_values(by='count', ascending=False).head(10)
top_tag_names = tags[tags['tag_id'].isin(top_tags.index)]['tag_name']

print("Top 10 Most Used Tags:")
print(top_tag_names)

# Get user input for selecting a tag
selected_tag = input("Please select a tag: ").strip().lower()  # Strip whitespace and convert to lowercase
print("Selected tag:", selected_tag)

# If no tag is selected, prompt the user to enter a tag
if not selected_tag:
    print("No tag selected. Please enter a tag.")
else:
    # Handle special characters and variations in tag names
    selected_tag = selected_tag.replace(r'[^a-zA-Z0-9\s]', '')

    # Check if the selected tag is in the list of top tag names
    selected_tag_id_series = tags[tags['tag_name'] == selected_tag]['tag_id']
    if not selected_tag_id_series.empty:
        selected_tag_id = selected_tag_id_series.values[0]
        print("Selected tag ID:", selected_tag_id)  # Print the selected tag ID

        # Filter book_tags DataFrame by selected tag ID
        filtered_books = book_tags[book_tags['tag_id'] == selected_tag_id]
        print("Filtered books:", filtered_books)  # Print the filtered books DataFrame

        if not filtered_books.empty:
            # Select a random book ID associated with the selected tag
            random_book_id = filtered_books.sample(1)['goodreads_book_id'].values[0]

            # Get the title of the selected book
            selected_book = books[books['book_id'] == random_book_id][['title']]
            print("Selected Book:")
            print(selected_book)
        else:
            print("No books were found associated with the selected tag.")
    else:
        print("Invalid tag. Please select a valid tag.")

# Print tag names from both DataFrames
print("Tag names in tags.csv:")
print(tags['tag_name'])
print("Tag names in book_tags.csv:")
print(book_tags['tag_id'])

# Print entries in book_tags.csv corresponding to the selected tag
print("Entries in book_tags.csv for selected tag:")
print(book_tags[book_tags['tag_id'] == selected_tag_id])

# Print entries in books.csv corresponding to the selected tag ID
print("Entries in books.csv for selected tag ID:")
print(books[books['book_id'].isin(filtered_books['goodreads_book_id'])])
```
